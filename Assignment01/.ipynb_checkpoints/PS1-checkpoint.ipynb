```python
#1. Flowchart
def purple_operator(x, y, z):
    return x + y - 10*z
    
def Print_values(a, b, c):
    # a > b 分支
    if a > b:
        if b > c:
            chosen_values = (a, b, c)
        else:
            if a > c:
                chosen_values = (a, c, b)
            else:
                chosen_values = (c, a, b)
        result = purple_operator(*chosen_values)
        print(f"Print_values({a}, {b}, {c})，Output = {result}")
        return result
    # a <= b 分支
    else:
        if b > c:
            # 课上老师说这里True就是直接 Stop，但是流程图的这个判断分支并没有指向椭圆STOP
            print(f"Print_values({a}, {b}, {c})：None")
            return None
        else:
            chosen_values = (c, b, a)
            result = purple_operator(*chosen_values)
            print(f"Print_values({a}, {b}, {c})，Output = {result}")
            return result

# Report your output when a = 5, b = 15, c = 10.
Print_values(5, 15, 10)  
```

```python
#2. Continuous celing function
import math
# memo = memoization ，基准情况 F(1) = 1 
memo = {1: 1} 

# 计算连续ceiling函数 F(x) = F(ceil(x/3)) + 2x ，然后使用记忆化来避免重复计算。
def F(x):
    if x in memo:
        return memo[x]
    
    # 计算 F(x)
    val = F(math.ceil(x / 3)) + 2 * x
    
    # 存储结果并返回
    memo[x] = val
    return val

test_list = [1, 2, 3, 5, 7, 10]
results = [F(x) for x in test_list]
print(f"测试列表 {test_list} 的 F(x) 结果:")
print(results)
```

```python
#3.Dice rolling

#3.1 Given 10 dice each with 6 faces, numbered from 1 to 6. Write a function Find_number_of_ways to find the number of ways to get sum x, defined as the sum of values on each face when all the dice are thrown.
import numpy as np

def get_dice_table(n_dice, n_faces):  # n_dice = 骰子数量，n_faces = 每个骰子的面数

    # 最大总和 = 骰子数量 × 面数
    max_sum = n_dice * n_faces
    
    # dp = Dynamic Programming , dp[i][j] 表示用 i 个骰子掷出总和为 j 的方式数
    # 初始化一个全零的二维数组，行 = 骰子数 + 1，列 = 最大总和 + 1
    dp = np.zeros((n_dice + 1, max_sum + 1), dtype=int)

    # 基准情况：0 个骰子总和为 0 的方式只有 1 zhong
    dp[0][0] = 1

    # 三层循环：
    for i in range(1, n_dice + 1):            # i = 骰子数量
        for j in range(i, i * n_faces + 1):    # j = 要计算的总和
            for k in range(1, n_faces + 1):     # k = 骰子可以掷出的点数（1~6）
                if j - k >= 0:
                    dp[i][j] += dp[i - 1][j - k]
    return dp


# 计算 10 个 6 面骰子的所有组合方式
N_DICE = 10
N_FACES = 6
dice_dp_table = get_dice_table(N_DICE, N_FACES)


# 定义函数：Find_number_of_ways，查找掷出总和为 x 的所有可能组合方式数量
def Find_number_of_ways(x):

    if 0 <= x <= N_DICE * N_FACES:
        return dice_dp_table[N_DICE][x]
    else:
        return 0


# 3.2 Count the number of ways for any x from 10 to 60, assign the number of ways to a list called Number_of_ways, so which x yields the maximum of Number_of_ways?
min_sum = N_DICE
max_sum = N_DICE * N_FACES

# 提取 10 到 60 的方式数量列表
Number_of_ways = dice_dp_table[N_DICE][min_sum:max_sum + 1]

# 找出最大值和他对应的总和值
max_ways = np.max(Number_of_ways)
x_with_max_ways = np.argmax(Number_of_ways) + min_sum

print(f"The sum x with the maximum number of ways = {x_with_max_ways}")
print(f"Maximum number of ways = {max_ways}")
```

```python
#4. Dynamic programming
import numpy as np
import matplotlib.pyplot as plt
#4.1 Write a function Random_integer to fill an array of N elements by randomly selecting integers from 0 to 10.

# 填充一个 N 元素的数组，随机整数范围为 1 到 10 (含 10) 。
def Random_integer(N):
    return np.random.randint(1, 11, size=N)

#4.2 Write a function Sum_averages to compute the sum of the average of all subsets of the array. For example, given an array of [1, 2, 3], you Sum_averages function should compute the sum of: average of [1], average of [2], average of [3], average of [1, 2], average of [1, 3], average of [2, 3], and average of [1, 2, 3].

#所有子集的平均值之和，根据 Sum(array) * (2^N - 1) / N 可以算出来。
def Sum_averages(array):
    N = len(array)
    if N == 0:
        return 0
    
    array_sum = np.sum(array)
    
    # (2^N - 1)
    # 使用 float 防止大数溢出
    factor = (np.power(2.0, N) - 1) / N
    
    return array_sum * factor

# 示例 [1, 2, 3]
array = [1, 2, 3]
print(f"Sum_averages result= {Sum_averages(array)}") 

#4.3 Call Sum_averages with N increasing from 1 to 100, assign the output to a list called Total_sum_averages. Plot Total_sum_averages, describe what do you see.

# N 从 1 到 100 
N_range = range(1, 101)
Total_sum_averages = []

for N in N_range:
    # 1. 生成随机数组
    random_arr = Random_integer(N)
    # 2. 计算 Sum_averages
    total_sum = Sum_averages(random_arr)
    # 3. 存入列表
    Total_sum_averages.append(total_sum)
    
#print(Total_sum_averages) 的输出结果很大 ，所也就不 print 出来了
#print(f"Sum_averages with N increasing from 1 to 100：{Total_sum_averages}")


# 绘制 Total_sum_averages 的结果
plt.figure(figsize=(10, 6))
plt.plot(N_range, Total_sum_averages)
plt.title('Total Sum Averages vs. N ')
plt.xlabel('N (Array Size)')
plt.ylabel('Total_sum_averages')
plt.grid(True)
plt.show()
```

```python
# 5. Path counting

#5.1 Create a matrix with N rows and M columns, fill the right-bottom corner and top-left corner cells with 1, and randomly fill the rest of matrix with integer 0 or 1.
import numpy as np
#p_one = 矩阵生成1的概率 ,mat = matrix
def create_matrix(N, M, p_one=0.5):
    mat = (np.random.rand(N, M) < p_one).astype(np.int64)
    mat[0, 0] = 1
    mat[N-1, M-1] = 1
    return mat


#5.2 Consider a cell marked with 0 as a blockage or dead-end, and a cell marked with 1 is good to go. Write a function Count_path to count total number of paths to reach the right-bottom corner cell from the top-left corner cell

#路径计算matrix[i,j]=1 可通行；=0 为障碍。
def count_path(matrix):
    N, M = matrix.shape
    paths = np.zeros((N, M), dtype=np.int64)

    # 起点
    paths[0, 0] = 1 
    if matrix[0, 0] == 1:
        paths[0, 0] = 1
    else:
        paths[0, 0] = 0

    # 第一列
    for i in range(1, N):
        if matrix[i, 0] == 1:
            paths[i, 0] = paths[i-1, 0]
        else:
            paths[i, 0] = 0

    # 第一行
    for j in range(1, M):
        if matrix[0, j] == 1:
            paths[0, j] = paths[0, j-1]
        else:
            paths[0, j] = 0

    # 其余格子
    for i in range(1, N):
        for j in range(1, M):
            if matrix[i, j] == 1:
                paths[i, j] = paths[i-1, j] + paths[i, j-1]
            else:
                paths[i, j] = 0
    #到达右下角的路径条数
    return int(paths[N-1, M-1])

#5.3 Let N = 10, M = 8, run Count_path for 1000 times, each time the matrix (except the right-bottom corner and top-left corner cells, which remain being 1) is re-filled with integer 0 or 1 randomly, report the mean of total number of paths from the 1000 runs.
N, M = 10, 8
mat = create_matrix(N, M, p_one=0.5)
print(mat)
paths = count_path(mat)
print(f"从 (0,0) 到 ({N-1},{M-1}) 的路径数: {paths}")

#重复实验 1000 次，求平均路径数
num_runs = 1000
path_counts = []
for _ in range(num_runs):
    mat = create_matrix(N, M, p_one=0.5)  # 每次新的随机矩阵
    path_counts.append(count_path(mat))

mean_paths = float(np.mean(path_counts))

print(f"N={N}, M={M}, 随机运行 {num_runs} 次,路径总数的平均值: {mean_paths:.3f}")

```
