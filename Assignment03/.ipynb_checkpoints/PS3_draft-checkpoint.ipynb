```python
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
ds = xr.open_dataset('200301_202006-C3S-L3_GHG-PRODUCTS-OBS4MIPS-MERGED-v4.3.nc')
#ÂèòÈáèËØªÂèñ
# Áî≤ÁÉ∑Êü±Âπ≥Âùá‰ΩìÁßØÂàÜÊï∞
xch4_da = ds["xch4"]
# Á∫¨Â∫¶
lat = ds["lat"].values
# ÁªèÂ∫¶
lon = ds["lon"].values
time = ds["time"].values


# ËΩ¨Êç¢‰∏∫ numpy Êï∞ÁªÑ
xch4 = xch4_da.values.astype(float)

# Â§ÑÁêÜÁº∫Â§±ÂÄºÔºàÂ§ß‰∫é 1e10 ÁöÑËÆ§‰∏∫ÊòØÁº∫Â§±Ôºâ
xch4[xch4 > 1e10] = np.nan

# Â∞ÜÊó∂Èó¥ËΩ¨Êç¢‰∏∫ pandas DatetimeIndex
time_index = pd.to_datetime(time)

# ÊèêÂèñÊúà‰ªΩÔºà1-12Ôºâ
month_array = np.array([t.month for t in time_index])
```

```python
#1.1 ËÆ°ÁÆó 12 ‰∏™ÊúàÁöÑÊ∞îÂÄôÊÄÅÔºàclimatologyÔºâÂπ∂ÁªòÂõæ
# ÂàõÂª∫‰∏Ä‰∏™ climatology Êï∞ÁªÑÔºö12 ‰∏™Êúà √ó Á∫¨Â∫¶ √ó ÁªèÂ∫¶
climatology = np.full((12, len(lat), len(lon)), np.nan)
# ËÆ°ÁÆóÊØè‰∏™ÊúàÁöÑÂπ≥Âùá
for m in range(1, 13):
    # ÊâæÂà∞ÊâÄÊúâÂ±û‰∫éÊúà‰ªΩ m ÁöÑÊó∂Èó¥Á¥¢Âºï
    mask = (month_array == m)
    
    # ÂèñÂá∫ËØ•ÊúàÂØπÂ∫îÁöÑÊâÄÊúâ xch4 Êï∞ÊçÆÔºåÂπ∂ÂØπÊó∂Èó¥Ê±ÇÂπ≥Âùá
    climatology[m-1, :, :] = np.nanmean(xch4[mask, :, :], axis=0)

# ÁªòÂà∂ 12 Âº†ÂõæÔºàÊØè‰∏™Êúà‰∏ÄÂº†Ôºâ
for m in range(1, 13):
    plt.figure(figsize=(6, 3))
    plt.imshow(climatology[m-1], origin="lower",
               extent=[lon.min(), lon.max(), lat.min(), lat.max()],
               aspect="auto")
    plt.colorbar(label="XCH4 (ppb)")
    plt.title("Methane Climatology - Month {}".format(m))
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.tight_layout()
    plt.show()
```

```python
#1.2 Plot globally-averaged methane from 2003-01 to 2020-06 as a time series. Describe your results. Check your plot with this one.
# ËΩ¨Êàê numpy Êï∞ÁªÑ
xch4 = xch4_da.values.astype(float)

# Áº∫Â§±ÂÄºfill_value:1e+20Â§ÑÁêÜ
xch4[xch4 > 1e20] = np.nan

# ÊääÊó∂Èó¥ËΩ¨Êç¢Êàê‰∏∫pdÁöÑDatetimeIndex
time_index = pd.to_datetime(time)

# ËÆ°ÁÆóÂÖ®ÁêÉÂπ≥ÂùáÔºàÂä†ÊùÉÂπ≥ÂùáÔºâ
# Á∫¨Â∫¶ÊùÉÈáçÔºöÈù¢ÁßØÊ≠£ÊØî‰∫é cos(Á∫¨Â∫¶)
lat_weight = np.cos(np.deg2rad(lat))   # ÂΩ¢Áä∂ (n_lat,)

# ÊääÁ∫¨Â∫¶ÊùÉÈáçÊâ©Â±ïÂà∞‰∫åÁª¥ÔºåÂíåÁªèÂ∫¶ÂåπÈÖçÔºöÂΩ¢Áä∂ (n_lat, n_lon)
weight_2d = lat_weight[:, None] * np.ones((1, len(lon)))

# Áî®Êù•Â≠òÊØè‰∏™Êó∂Èó¥ÁÇπÁöÑÂÖ®ÁêÉÂπ≥ÂùáÂÄº
global_mean = np.full(len(time_index), np.nan)

# Âæ™ÁéØÊØè‰∏Ä‰∏™Êó∂Èó¥Ê≠•ÔºåËÆ°ÁÆóÂä†ÊùÉÂπ≥Âùá
for i in range(len(time_index)):
    data = xch4[i, :, :]   # ÂΩìÂâçÊó∂Èó¥ÁöÑ 2D Âú∫ÔºåÂΩ¢Áä∂ (n_lat, n_lon)
    
    # ÊúâÊïàÂÄºÊé©Á†ÅÔºà‰∏çÊòØ NaN ÁöÑÂú∞ÊñπÔºâ
    mask = np.isfinite(data)
    
    # Â¶ÇÊûúËøô‰∏ÄÊó∂ÂàªÊï∞ÊçÆÂÖ®ÊòØ NaNÔºåÂ∞±Ë∑≥Ëøá
    if np.sum(mask) == 0:
        global_mean[i] = np.nan
        continue
    
    # Êã∑Ë¥ù‰∏Ä‰ªΩÊùÉÈáçÔºåÂπ∂Âú®Êó†Êïà‰ΩçÁΩÆËÆæ‰∏∫ 0
    w = weight_2d.copy()
    w[~mask] = 0.0
    
    # ÂΩí‰∏ÄÂåñÊùÉÈáçÔºàÊâÄÊúâÊúâÊïàÊ†ºÁÇπÁöÑÊùÉÈáçÂíå‰∏∫ 1Ôºâ
    w = w / np.nansum(w)
    
    # ËÆ°ÁÆóÂä†ÊùÉÂπ≥ÂùáÔºà‚àë data * wÔºâ
    global_mean[i] = np.nansum(data * w)

# Êó∂Èó¥Â∫èÂàóÂõæ

plt.figure(figsize=(10, 4))
plt.plot(time_index, global_mean, marker='o', markersize=2,linewidth=1.0)

plt.xlabel("Time")
plt.ylabel("Global mean XCH4") 
plt.title("Global Mean Column-Averaged Methane (2003-01 to 2020-06)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#3 
# 2. ÊâæÂà∞ÊúÄËøëÁöÑÁΩëÊ†ºÁÇπÔºà15¬∞S, 150¬∞WÔºâ
target_lat = -15.0   # ÁõÆÊ†áÁ∫¨Â∫¶ÔºàÂçóÁ∫¨15Â∫¶Ôºâ
target_lon = -150.0  # ÁõÆÊ†áÁªèÂ∫¶ÔºàË•øÁªè150Â∫¶ÔºåÂØπÂ∫î -150Ôºâ

# ÊâæÂà∞Á¶ªÁõÆÊ†áÁÇπÊúÄËøëÁöÑÁΩëÊ†ºÁ¥¢Âºï
idx_lat = np.argmin(np.abs(lat - target_lat))
idx_lon = np.argmin(np.abs(lon - target_lon))

print("ÂÆûÈôÖ‰ΩøÁî®ÁöÑÁΩëÊ†ºÁÇπÔºö")
print("lat =", lat[idx_lat], "deg")
print("lon =", lon[idx_lon], "deg")

# ÂèñÂá∫ËØ•ÁÇπÁöÑÊó∂Èó¥Â∫èÂàó
series = xch4[:, idx_lat, idx_lon]

print("Ëøô‰∏™ÁÇπÁöÑÊó∂Èó¥Â∫èÂàóÈïøÂ∫¶:", len(series))

# 3. ËÆ°ÁÆóËØ•ÁÇπÁöÑÊúàÂπ≥ÂùáÊ∞îÂÄôÊÄÅÔºà1-12 ÊúàÔºâ

clim_point = np.full(12, np.nan)  # Â≠òÊîæ 12 ‰∏™ÊúàÁöÑÂπ≥ÂùáÂÄº

for m in range(1, 13):
    # ÊâæÂá∫ÊâÄÊúâÂ±û‰∫éÁ¨¨ m ÊúàÁöÑÊï∞ÊçÆ
    mask = (month_array == m)
    data_month = series[mask]
    
    # Âè™ÂØπÊúâÊúâÊïàÊï∞ÊçÆÁöÑÊúà‰ªΩÊ±ÇÂπ≥Âùá
    valid = np.isfinite(data_month)
    if np.sum(valid) == 0:
        print("Month", m, "Âú®Ëøô‰∏™ÁÇπÊ≤°ÊúâÊúâÊïàÊï∞ÊçÆÔºåclimatology ‰øùÊåÅ NaN")
        continue
    
    clim_point[m-1] = np.mean(data_month[valid])

print("ËØ•ÁÇπ 12 ‰∏™Êúà climatologyÔºö")
print(clim_point)

# 4. ÂéªÂ≠£ËäÇÂåñÔºàÂéüÂ∫èÂàóÂáèÂéªÂØπÂ∫îÊúà‰ªΩ climatologyÔºâ

deseason = np.full_like(series, np.nan, dtype=float)

for i in range(len(series)):
    value = series[i]
    if not np.isfinite(value):
        # Â¶ÇÊûúÂéüÂßãÂÄºÂ∞±ÊòØ NaNÔºåÂ∞±‰øùÊåÅ NaN
        continue
    
    m = month_array[i]       # ÂΩìÂâçÊó∂Èó¥Â±û‰∫éÂá†Êúà‰ªΩÔºà1-12Ôºâ
    clim_val = clim_point[m-1]
    
    if np.isfinite(clim_val):
        deseason[i] = value - clim_val
    else:
        # Â¶ÇÊûúËøô‰∏™ÊúàÁöÑ climatology ÊòØ NaNÔºåÂ∞±‰øùÊåÅ NaN
        deseason[i] = np.nan

#ÂéªÂ≠£ËäÇÂåñÂêéÁöÑÊó∂Èó¥Â∫èÂàó

plt.figure(figsize=(10, 4))

plt.plot(time_index, deseason, marker="o",ms = 2.0, linewidth=1.0)
plt.axhline(0, color="gray", linestyle="--")

plt.xlabel("Time")
plt.ylabel("Deseasonalized XCH4")
plt.title("Deseasonalized Methane at (~15¬∞S, 150¬∞W), 2003-01 to 2020-06")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#2.1

ds2 = xr.open_dataset( "NOAA_NCDC_ERSST_v3b_SST.nc")
#ÂèòÈáè
sst_da = ds2["sst"]
lat = ds2["lat"].values
lon = ds2["lon"].values
time = ds2["time"].values  
sst = sst_da.values.astype(float)

#Ni√±o 3.4 Âå∫Âüü
lat_mask = (lat >= -5.0) & (lat <= 5.0)
lon_mask = (lon >= 190.0) & (lon <= 240.0)

lat_sel = lat[lat_mask]
lon_sel = lon[lon_mask]

print("Ni√±o3.4 Á∫¨Â∫¶ÁÇπ:", lat_sel)
print("Ni√±o3.4 ÁªèÂ∫¶ËåÉÂõ¥:", lon_sel[0], "to", lon_sel[-1])

# Âèñ Ni√±o3.4 Âå∫ÂüüÂØπÂ∫îÁöÑ SST (time, nlat, nlon)
sst_region = sst[:, lat_mask, :][:, :, lon_mask]
print("Ni√±o3.4 Âå∫Âüü SST ÂΩ¢Áä∂:", sst_region.shape)


# 3. ÂØπ Ni√±o 3.4 Âå∫ÂüüÂÅöÈù¢ÁßØÂä†ÊùÉÂπ≥ÂùáÔºàÂæóÂà∞ 1D Êó∂Èó¥Â∫èÂàóÔºâ
# Á∫¨Â∫¶ÊùÉÈáçÔºöcos(lat)
lat_weights = np.cos(np.deg2rad(lat_sel))     # (nlat,)
# Êâ©Â±ïÊàê‰∫åÁª¥ (nlat, nlon)
weight_2d = lat_weights[:, None] * np.ones((1, len(lon_sel)))
# ÂÖàÊï¥‰ΩìÂΩí‰∏ÄÂåñ‰∏ÄÊ¨°
weight_2d = weight_2d / np.nansum(weight_2d)

# ÁªìÊûúÔºöNi√±o3.4 ÊØèÊúàÂπ≥Âùá SST Â∫èÂàó
nino_ts = np.full(sst_region.shape[0], np.nan)

for i in range(sst_region.shape[0]):
    field = sst_region[i, :, :]   # ÂΩìÊúàÂú® Ni√±o3.4 Âå∫ÂüüÁöÑÂú∫
    mask_valid = np.isfinite(field)
    
    if np.sum(mask_valid) == 0:
        nino_ts[i] = np.nan
        continue
    
    # ÈíàÂØπÂΩìÊúàÊúâÊïàÊ†ºÁÇπÈáçÊñ∞ÂΩí‰∏ÄÂåñÊùÉÈáç
    w = weight_2d.copy()
    w[~mask_valid] = 0.0
    w = w / np.nansum(w)
    
    nino_ts[i] = np.nansum(field * w)

print("Ni√±o3.4 Êó∂Èó¥Â∫èÂàóÈïøÂ∫¶:", len(nino_ts))

# 4. ËÆ°ÁÆóÊúàÊ∞îÂÄôÊÄÅÔºàclimatologyÔºâÂπ∂ÂæóÂà∞ÂºÇÂ∏∏ÔºàanomaliesÔºâ

time_index = pd.to_datetime(time)
month_array = np.array([t.month for t in time_index])

# 12 ‰∏™ÊúàÁöÑÊ∞îÂÄôÊÄÅ
climatology = np.full(12, np.nan)

for m in range(1, 13):
    mask_month = (month_array == m)
    data_month = nino_ts[mask_month]
    
    valid = np.isfinite(data_month)
    climatology[m-1] = np.mean(data_month[valid])

print("Ni√±o3.4 Âå∫Âüü 12 ‰∏™Êúà climatology:")
print(climatology)

# ËÆ°ÁÆóÂºÇÂ∏∏ÔºöÂΩìÂâçÂÄº - ÂΩìÊúàÊ∞îÂÄôÊÄÅ
nino_anom = np.full_like(nino_ts, np.nan, dtype=float)

for i in range(len(nino_ts)):
    if not np.isfinite(nino_ts[i]):
        continue
    m = month_array[i]
    nino_anom[i] = nino_ts[i] - climatology[m-1]

# ÁÆÄÂçïÁîª‰∏Ä‰∏ãÂºÇÂ∏∏Êó∂Èó¥Â∫èÂàó
plt.figure(figsize=(10, 4))
plt.plot(time_index, nino_anom, linewidth=0.8)
plt.axhline(0.0, color="black", linestyle="--")
plt.xlabel("Time")
plt.ylabel("Ni√±o3.4 SST Anomaly (¬∞C)")
plt.title("Ni√±o3.4 SST Monthly Anomalies")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#ËÆ°ÁÆó 3 ‰∏™ÊúàÊªëÂä®Âπ≥ÂùáÔºàNi√±o 3.4 ÊåáÊï∞Ôºâ
nino_index = np.full_like(nino_anom, np.nan, dtype=float)

for i in range(len(nino_anom)):
    # ÊªëÂä®Á™óÂè£ÁöÑÂºÄÂßãÂíåÁªìÊùüÁ¥¢Âºï
    start = max(0, i - 1)
    end = min(len(nino_anom), i + 2)  # Python ÂàáÁâáÂè≥Á´Ø‰∏çÂåÖÂê´
    
    window = nino_anom[start:end]
    valid = np.isfinite(window)
    
    # Ëá≥Â∞ëÊúâ 2 ‰∏™ÊúâÊïàÁÇπÂÜçÊ±ÇÂπ≥Âùá
    if np.sum(valid) >= 2:
        nino_index[i] = np.mean(window[valid])

plt.figure(figsize=(10, 5))

# Ê≠£ÂºÇÂ∏∏ (El Ni√±o) ‚Äî Á∫¢Ëâ≤
plt.fill_between(time_index, nino_index, 0, where=nino_index >= 0,  color='red', alpha=0.6, label="El Ni√±o (‚â• 0¬∞C)")

# Ë¥üÂºÇÂ∏∏ (La Ni√±a) ‚Äî ËìùËâ≤
plt.fill_between(time_index, nino_index, 0, where=nino_index < 0, color='blue', alpha=0.6, label="La Ni√±a (< 0¬∞C)")

# ÁîªÂá∫ 3 ‰∏™ÊúàÊªëÂä®Âπ≥ÂùáÊõ≤Á∫ø
plt.plot(time_index, nino_index, color="black", linewidth=1, label="3-month Running Mean")

# ÂàÜÁïåÁ∫ø 0¬∞C
plt.axhline(0, color="black", linewidth=1)

# 3. ÂõæÂÉèËÆæÁΩÆ

plt.title("Ni√±o 3.4 Index (3-month Running Mean)", fontsize=14)
plt.ylabel("SST Anomaly (¬∞C)")
plt.xlabel("Time")
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(loc="upper left")
plt.tight_layout()

plt.show()
```

```python
#3.1

import xarray as xr
data_path = '198001_202401_MERRA2_BCCMASS_Data/*.nc4'
bc_ds = xr.open_mfdataset(data_path, combine='by_coords', parallel=True)

# ËÆ°ÁÆóÂÖ®ÁêÉÂπ≥ÂùáÊó∂Èó¥Â∫èÂàó (Global Mean Time Series) Ôºå‰ΩøÁî® cos(lat) ËøõË°åÂä†ÊùÉÂπ≥Âùá„ÄÇ
weights = np.cos(np.deg2rad(bc_ds.lat))
weights.name = "weights"

# ËÆ°ÁÆóÂä†ÊùÉÂêéÁöÑÂÖ®ÁêÉÂπ≥ÂùáÂÄºÔºàËá™Âä®ÂπøÊí≠„ÄÅÂΩí‰∏ÄÂåñ„ÄÅÂøΩÁï•Áº∫Â§±ÂÄºÔºâ
bcc_global = bc_ds['BCCMASS'].weighted(weights).mean(dim=("lat", "lon"))

# ËÆ°ÁÆóÊ∞îÂÄôÊÄÅÔºàÊØè‰∏™ÊúàÁöÑÂ§öÂπ¥Âπ≥ÂùáÂÄºÔºâ
climatology = bcc_global.groupby("time.month").mean("time")

# ÂéªÈô§Â≠£ËäÇÂæ™ÁéØÔºåËÆ°ÁÆó Anomalies ÔºàÂéüÂßãÂÄº - Ê∞îÂÄôÊÄÅÔºâ
bcc_anomalies = bcc_global.groupby("time.month") - climatology

# ÁîªÂõæ
fig, ax = plt.subplots(figsize=(12, 6))
bcc_anomalies.plot(ax=ax, label='De-seasonalized (Anomalies)', color='red', linewidth=1.5)


plt.title('Deseasonalized Global Mean BC Column Mass Density')
plt.xlabel('Time')
plt.ylabel('BC Column Mass Density Anomaly\n(kg/m^2, deseasonalized)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# 3.2

import matplotlib.pyplot as plt
import cartopy.crs as ccrs

# ÂèñÂèòÈáè
bc = bc_ds['BCCMASS']
lat = bc_ds['lat']
lon = bc_ds['lon']
time = bc_ds['time']

# Fig.1.ÂÖ®ÁêÉÂπ≥ÂùáBCÂéüÂßãÊó∂Èó¥Â∫èÂàó
plt.figure(figsize=(10,6))
bcc_global.plot(label="Global Mean BC")
plt.title("Global Mean BC Column Mass Density (1980‚Äì2025)")
plt.ylabel("kg m$^{-2}$")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()


# Fig.2.ÂÖ®ÁêÉBCÂπ≥ÂùáÁ©∫Èó¥ÂàÜÂ∏ÉÂõæ
# ËÆ°ÁÆóÊó∂Èó¥Âπ≥ÂùáÔºåÊ∂àÈô§Êó∂Èó¥Áª¥Â∫¶ÁöÑÊ≥¢Âä®ÔºåÊü•ÁúãÈïøÊúüÂ≠òÂú®ÁöÑÊéíÊîæÁÉ≠ÁÇπ
bcc_mean = bc.mean(dim='time')

fig = plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_xticks(np.arange(-180,181,60), crs=ccrs.PlateCarree())
ax.set_yticks(np.arange(-90,91,30), crs=ccrs.PlateCarree())

bcc_mean.plot(ax=ax, transform=ccrs.PlateCarree() , cbar_kwargs={'label': 'kg m-2'} , cmap='OrRd')

plt.title("Long-term Mean Black Carbon Mass Density (1980-2025)")

plt.show()



# Fig.3. BCÂÖ®ÁêÉÂ≠£ËäÇÊÄßÂæ™ÁéØÊó∂Èó¥Â∫èÂàóÂõæ
climatology = bcc_global.groupby("time.month").mean("time")


plt.figure(figsize=(6,4), dpi=150)
plt.plot(np.arange(1,13), climatology, marker='o')
plt.xticks(np.arange(1,13))
plt.title("Seasonal Cycle of Global Mean BC")
plt.xlabel("Month")
plt.ylabel("kg m$^{-2}$")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()



# Fig.4.BCÁ∫¨ÂêëÂπ≥ÂùáÂàÜÂ∏ÉÈù¢Âõæ
#Â±ïÁ§∫‰∏çÂêåÁ∫¨Â∫¶Â∏¶ÁöÑÂπ≥ÂùáÊ±°ÊüìÊ∞¥Âπ≥
# Âü∫‰∫é bc_ds ÁöÑÊó∂Èó¥Âπ≥ÂùáÂú∫ÔºåÂÜçËøõË°åÁªèÂ∫¶Âπ≥Âùá
bc_zonal_profile = bc.mean(dim='time').mean(dim='lon')   # ÁªìÊûúÁª¥Â∫¶Ôºö(lat,)

fig, ax = plt.subplots(figsize=(6, 4), dpi=150)

# x ËΩ¥ÔºöÈªëÁ¢≥ÂØÜÂ∫¶Ôºõy ËΩ¥ÔºöÁ∫¨Â∫¶
ax.plot(bc_zonal_profile, lat, color='black', linewidth=1.5)
# Ëµ§ÈÅìÂèÇËÄÉÁ∫ø
ax.axhline(0, color='grey', linestyle='--')
ax.set_ylim(-90, 90)
ax.set_title("Zonal Mean Profile (Time Averaged)")
ax.set_xlabel("Black Carbon Column Density (kg m$^{-2}$)")
ax.set_ylabel("Latitude")
ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()



# Fig.5. ÈúçÂ§´Ëé´ÂãíÂõæ (Time-Lat Hovm√∂ller)
#ËÆ°ÁÆóÁªèÂ∫¶Âπ≥Âùá
bc_zonal = bc.mean(dim="lon")     # (time, lat)

fig, ax = plt.subplots(figsize=(12, 6))
# xËΩ¥‰∏∫Êó∂Èó¥ÔºåyËΩ¥‰∏∫ÂîØÁã¨   
bcc_zonal.plot(x='time', y='lat', robust=True, cmap='Spectral_r')
plt.title("Hovm√∂ller Diagram: Time‚ÄìLatitude BC")
plt.xlabel("Time")
plt.ylabel("Latitude")
plt.tight_layout()
plt.show()
```
