```python
import pandas as pd
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import os
```

```python
# 1.1 Global Earthquakes
df = pd.read_csv("usgs_earthquakes.csv")
df_top50 = df.sort_values(by="mag", ascending=False).head(50)

#  Robinson 
fig = plt.figure(figsize=(12, 6), dpi=150) 
proj = ccrs.Robinson(central_longitude=180)
ax = fig.add_subplot(1, 1, 1, projection=proj)
ax.set_global()

# 
ax.stock_img() 
ax.coastlines(linewidth=0.5) 

# top50
lons = df_top50["longitude"]
lats = df_top50["latitude"]
mags = df_top50["mag"]

sc = ax.scatter(
    lons, lats,
    c=mags,
    cmap="Reds",
    vmin=mags.min(), 
    vmax=mags.max(),
    edgecolors="black",
    transform=ccrs.PlateCarree(), 
    zorder=3
)

cbar = fig.colorbar(sc, ax=ax, shrink=0.8, pad=0.04)
cbar.set_label("Magnitude")

ax.set_title("Top 50 Earthquakes of 2014", fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# 2.1 Global black carbon analysis
# 
data_path = '198001_202401_MERRA2_BCCMASS_Data/*.nc4'
bc_ds = xr.open_mfdataset(data_path, combine='by_coords', parallel=True)
bc = bc_ds['BCCMASS']

#  19802024 
bc_mean = bc.mean(dim='time').compute()

# 
lats = bc_mean['lat']
lons = bc_mean['lon']
lon2d, lat2d = np.meshgrid(lons, lats)

# 
units = bc_mean.attrs.get('units', 'kg m$^{-2}$')

fig = plt.figure(figsize=(12, 6), dpi=150)

# PlateCarree
proj_global = ccrs.PlateCarree(central_longitude=0)
ax = fig.add_subplot(1, 1, 1, projection=proj_global)
ax.set_global()

# 
ax.stock_img()
ax.coastlines(linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linewidth=0.3)

# 
pc = ax.pcolormesh(
    lon2d, lat2d, bc_mean,
    transform=ccrs.PlateCarree(),
    cmap='OrRd'
)

# Colorbar
cbar = plt.colorbar(pc, ax=ax, orientation='vertical', pad=0.04, shrink= 0.8)
cbar.set_label(f'BC column mass ({units})')
cbar.ax.tick_params(labelsize=8)

# Gridlines + labels
gl = ax.gridlines(draw_labels=True, linestyle='--', linewidth=0.5, alpha=0.7)
gl.top_labels = False
gl.right_labels = False
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# 
bc_np = bc_mean.values
j, i = np.unravel_index(np.nanargmax(bc_np), bc_np.shape)
max_lat = float(lats[j])
max_lon = float(lons[i])

# bc
ax.plot(
    max_lon, max_lat,
    marker='o', markersize=12, markerfacecolor='none',markeredgecolor='blue',linewidth=2,linestyle='None',
    transform=ccrs.PlateCarree(),
    label='Max BC Column Mass Location'
)

# 
ax.legend(loc='lower left', frameon=True, edgecolor='gray',  bbox_to_anchor=(0 , 0),  markerscale=0.75,  fancybox=False)

#
ax.text(
    0.01, -0.1,
    f'Max BC Location: {max_lat:.2f}N, {max_lon:.2f}E',
    transform=ax.transAxes,
    ha='left', va='bottom',
    fontsize=10,
    bbox=dict(
        facecolor='white',
        edgecolor='black',
        boxstyle='square,pad=0.3', 
        alpha=0.8
    )
)

# 
ax.set_title('Global Mean BC Column Mass (19802024, MERRA-2)', fontsize=16, pad=10)

plt.tight_layout()
plt.show()
```

```python
# 2.2 BC 
# 60150E, 060N
bc_reg = bc_mean.sel(lon=slice(60, 150), lat=slice(0, 60))
lats_r = bc_reg['lat']
lons_r = bc_reg['lon']
lon2d_r, lat2d_r = np.meshgrid(lons_r, lats_r)

fig = plt.figure(figsize=(8, 6), dpi=150)

# Mercator
proj_reg = ccrs.Mercator(central_longitude=110)
ax = fig.add_subplot(1, 1, 1, projection=proj_reg)

# 
ax.set_extent([60, 150, 0, 60], crs=ccrs.PlateCarree())

# 
ax.add_feature(cfeature.OCEAN, facecolor='0.85')
ax.add_feature(cfeature.LAND, facecolor='0.9')
ax.coastlines(linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linewidth=0.4)

# 
pc = ax.pcolormesh(lon2d_r, lat2d_r, bc_reg, transform=ccrs.PlateCarree(), cmap='YlOrBr', vmin=0, vmax=5e-6)

# 
cbar = plt.colorbar(pc, ax=ax, orientation='vertical', pad=0.04, shrink=0.88)
cbar.set_label(f'BC column mass ({units})')

# 
gl = ax.gridlines(draw_labels=True, linestyle='--', linewidth=0.5, alpha=0.7)
gl.top_labels = False
gl.right_labels = False

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

#  
bc_reg_np = bc_reg.values
j_r, i_r = np.unravel_index(np.nanargmax(bc_reg_np), bc_reg_np.shape)
max_lat_r = float(lats_r[j_r])
max_lon_r = float(lons_r[i_r])

# 
ax.plot(
    max_lon_r, max_lat_r,
    marker='o', markersize=15,
    markerfacecolor='none',
    markeredgecolor='blue',
    markeredgewidth=1.5,
    linestyle='None',
    transform=ccrs.PlateCarree(),
    label='Max BC Region in East Asia'
)

# 
ax.text(
    0.015, 0.94,
    'East Asia (60150E, 060N)\n19802024 mean BC',
    transform=ax.transAxes,
    fontsize=8,
    bbox=dict(
        facecolor='white',
        edgecolor='black',
        boxstyle='square,pad=0.3',
        alpha=0.8
    )
)

# 
legend = ax.legend(loc='lower left', frameon=True, edgecolor='gray', markerscale=0.6, fontsize=8,)

ax.set_title('East Asia BC Column Mass (19802024, MERRA-2)', fontsize=12, pad=10)

plt.tight_layout()
plt.show()
```
